```python
import re

def extract_postal_code(address):
    if address is None or not isinstance(address, str):
        return None, None

    prefixes = ['E', 'EC', 'N', 'NW', 'SE', 'SW', 'W', 'WC', 'BR', 'CM', 'CR', 'DA', 'EN', 'HA', 'IG', 'SL', 'TN', 'KT',
                'RM', 'SM', 'TW', 'UB', 'WD', 'LU', 'RH']

    pattern = r'\b(?:' + '|'.join(prefixes) + r')\d+[A-Z]?\b'

    match = re.search(pattern, address)

    if match:
        post_code=match.group(0)
        lis=address.split(post_code)
        if len(lis)>1:
          st_lis=lis[1].split(',')
          if len(st_lis)>1:
            return post_code, st_lis[0].strip()
          else:
            return post_code, lis[1].strip()
        else:
          return post_code, None
    else:
        return None, None  # Return None for both values if no match


final_df['Pick_up_postcode'],final_df['Pick_up_streetcode'] = zip(*final_df['Pick-up address'].apply(extract_postal_code))
final_df['Drop_off_postcode'], final_df['Drop_off_streetcode'] = zip(*final_df['Drop off address'].apply(extract_postal_code))
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.express as px

def plot_driver_performance_plotly(driver_hourly_pivot, driver_names):
    fig = go.Figure()

    # Define hours (0 to 23) as the x-axis values
    hours = list(range(24))

    # Add traces for each driver, but make them initially invisible
    for driver_name in driver_names:
        driver_data = driver_hourly_pivot[driver_hourly_pivot['full_name'] == driver_name]

        if driver_data.empty:
            print(f"Error: No data found for driver '{driver_name}'.")
            continue

        mean_earnings = driver_data.filter(regex='Mean Earnings_').values.flatten()
        mean_distance = driver_data.filter(regex='Mean Distance travel_').values.flatten()


        earnings_trace = go.Scatter(
            x=hours,
            y=mean_earnings,
            mode='lines',
            name=f'{driver_name} - Mean Earnings',
            visible=False
        )

        distance_trace = go.Scatter(
            x=hours,
            y=mean_distance,
            mode='lines',
            name=f'{driver_name} - Mean Distance Travel',
            visible=False
        )

        fig.add_trace(earnings_trace)
        fig.add_trace(distance_trace)

    dropdown_buttons = []
    for i, driver_name in enumerate(driver_names):
        dropdown_buttons.append(
            dict(
                args=[{'visible': [False] * len(fig.data)}],  # Hide all traces first
                label=driver_name,
                method="update"
            )
        )
        dropdown_buttons[-1]['args'][0]['visible'][i*2] = True  # Show earnings trace
        dropdown_buttons[-1]['args'][0]['visible'][i*2+1] = True  # Show distance trace

    # Layout for the plot
    fig.update_layout(
        updatemenus=[
            dict(
                buttons=dropdown_buttons,
                direction="down",
                showactive=True,
                x=0.17,
                xanchor="left",
                y=1.15,
                yanchor="top"
            )
        ],
        title="Driver Performance (Mean Earnings & Mean Distance Travel) per Hour",
        xaxis_title="Hour",
        yaxis_title="Value",
    )

    fig.show()

driver_names = list(driver_hourly_pivot['full_name'].unique())
plot_driver_performance_plotly(driver_hourly_pivot, driver_names)
```

```python
import re

# Function to sanitize and format DMS string
def clean_and_format_dms(dms_str):
    # Remove unwanted characters and normalize spacing
    dms_str = dms_str.replace('Â', '').strip()
    dms_str = re.sub(r"[']+", '"', dms_str)  # Replace single quotes with double quotes
    dms_str = re.sub(r"\s+", " ", dms_str)  # Normalize spaces

    # Regex pattern to extract degrees, minutes, seconds, and direction
    dms_pattern = re.compile(r'(\d+)[° ]+(\d+)?[\'"]?\s*(\d+(?:\.\d+)?)?[\'"]?\s*([NSEW])')
    match = dms_pattern.search(dms_str)

    if not match:
        raise ValueError(f"Invalid format for DMS: {dms_str}")

    # Extract degrees, minutes, seconds, and direction
    degrees = match.group(1)  # Degrees
    minutes = match.group(2) if match.group(2) else '0'  # Minutes (default to 0 if missing)
    seconds = match.group(3) if match.group(3) else '0'  # Seconds (default to 0 if missing)
    direction = match.group(4)  # Direction (N, S, E, W)

    # Format into the desired "51° 30' 31.1" N" format
    formatted_dms = f'{degrees}° {minutes}\' {seconds}" {direction}'
    return formatted_dms, degrees, minutes, seconds, direction


# Function to convert DMS to Decimal Degrees
def dms_to_decimal(degrees, minutes, seconds, direction):
    # Convert to decimal degrees
    decimal_degrees = float(degrees) + (float(minutes) / 60) + (float(seconds) / 3600)

    # If the direction is South or West, make the value negative
    if direction in ['S', 'W']:
        decimal_degrees = -decimal_degrees

    return decimal_degrees


# Function to process CSV file and standardize DMS format
def process_and_convert_dms(df,lat,lon):
    lat=lat
    lon=lon
    # Check if lat and lon columns exist in the DataFrame
    if lat not in df.columns or lon not in df.columns:
        raise ValueError("Columns lat and lon are missing from the CSV file.")

    # Initialize lists for decimal values
    decimal_latitudes = []
    decimal_longitudes = []

    # Clean and format each entry in lat and lon columns
    for index, row in df.iterrows():
        # Process latitude
        formatted_latitude, degrees_lat, minutes_lat, seconds_lat, direction_lat = clean_and_format_dms(row[lat])
        df.at[index, lat] = formatted_latitude  # Update DataFrame with formatted DMS
        decimal_latitudes.append(dms_to_decimal(degrees_lat, minutes_lat, seconds_lat, direction_lat))

        # Process longitude
        formatted_longitude, degrees_lon, minutes_lon, seconds_lon, direction_lon = clean_and_format_dms(row[lon])
        df.at[index, lon] = formatted_longitude  # Update DataFrame with formatted DMS
        decimal_longitudes.append(dms_to_decimal(degrees_lon, minutes_lon, seconds_lon, direction_lon))

    # Add new columns for decimal degrees
    df['end_latitude_decimal'] = decimal_latitudes
    df['end_longitude_decimal'] = decimal_longitudes
    return df
final_df=process_and_convert_dms(final_df,"end_latitude","end_longitude")
final_df=process_and_convert_dms(final_df,"start_latitude","start_longitude")
```
